```python
import yfinance as yf
import pandas as pd
```

```python
tickers = {
    "TSLA": "Tesla",
    "F": "Ford",
    "BMWYY": "BMW",        
    "VWAGY": "Volkswagen", 
    "HYMTF": "Hyundai"     
}

merged_data = pd.DataFrame()

for ticker, prefix in tickers.items():
    data = yf.download(ticker, start="2019-01-01", end="2024-11-01")
    data = data.add_prefix(f"{prefix}_")
    if merged_data.empty:
        merged_data = data
    else:
        merged_data = pd.merge(merged_data, data, left_index=True, right_index=True)

merged_data.to_csv("tesla_ford_bmw_volkswagen_hyundai_full.csv")
```
